# Descriptive statistics

```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns

# Loading data
df = pd.read_pickle('STING_DB.pkl')
df.head()
```

## Clinical features

```python
# Sex
df.groupby('sex')['id'].nunique().to_frame()
```

```python
# Age at diagnosis, years
df.groupby('id').agg({'age_dx':max}).describe().astype('int')
```

```python
# Follow-up, months
df['fu_months'] = (df['fu_days']/30)
df.groupby('id').agg({'fu_months':max}).describe().astype(int)
```
