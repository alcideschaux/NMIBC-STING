```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats

sns.set(font_scale=1.25, rc={'figure.figsize':[12,8]})
sns.set_style('white')
```

```python
df = pd.read_pickle('STING_DB.pkl')
df.head()
```

# Naïve-treatment cases

```python
# Selecting naïve-treatment cases
df.query('turb==0', inplace=True)

# Number of naïve-treatment cases
df['sp'].nunique()
```

```python
group = 'pt'
df = df.astype({'pt':'category'})
df['pt'].cat.reorder_categories(['Tis','Ta','T1'], ordered=True, inplace=True)
```

```python
marker = 'sting_tumor'
display(df.groupby(group)[marker].describe().round())

p = scipy.stats.kruskal(*[data[marker].values for name, data in df.groupby(group)], nan_policy='omit')[1]
sns.boxplot(x=group, y=marker, data=df)
plt.annotate(text='P value = {:.1e}'.format(p), xy=(0,100), ha='center', va='center')
plt.show()
```
